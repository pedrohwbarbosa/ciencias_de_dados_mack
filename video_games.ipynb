```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Leitura dos dados
df = pd.read_csv(r'C:\Users\pedro\Desktop\dataset\vgsales.csv')
```

```python
# 2. Print que mostre a quantidade de objetos e atributos
print("Quantidade de objetos e atributos:")
print(f"Linhas: {df.shape[0]}, Colunas: {df.shape[1]}")
```

```python
# 3. Print que detalhe os tipos de dados
print("\nTipos de dados de cada atributo:")
print(df.dtypes)
```

```python
# 4. Print com análises estatísticas dos atributos (média/mediana, dispersão e etc)
print("\nAnálise estatística dos atributos numéricos:")
print(df.describe())
```

```python
# 5. Escolha de duas ou mais variáveis numéricas para plotar boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['NA_Sales', 'Global_Sales']])
plt.title('Boxplot das Vendas na América do Norte e Vendas Globais')
plt.show()
```

```python
# 6. Escolha de uma ou mais variáveis categóricas para plotar gráfico de barras
plt.figure(figsize=(12, 6))
sns.countplot(y='Genre', data=df, order=df['Genre'].value_counts().index)
plt.title('Contagem de Jogos por Gênero')
plt.show()
```
